## 1. My lane (or freestyle) and why

**Selected Lane**: Lane 2: Refresh / Content Opportunity Scoring

Content decay is a primary challenge in organic search visibility. Over time, once-high-ranking pages slip in positions and clicks due to search engine updates, competitor adjustments, or shifting audience interest. For platforms managing hundreds or thousands of pages, auditing and updating every page is financially and operationally impossible. By implementing a machine learning scoring system, we can rank pages by their likelihood of decline and their search exposure. This prioritizes human editorial resources for the highest-impact refresh opportunities, reversing traffic loss before it severely impacts revenue.

## 2. The question: decision, action, cost of a wrong call

- **What decision does this improve?** Which decaying or opportunity-rich page should a FlyRank editor review and refresh first to maximize organic traffic recovery?
- **Who acts on it?** The FlyRank content editorial and optimization team.
- **What is their action?** Editors manually inspect the flagged page, update outdated facts, expand the content depth, improve layout structure, and republish the article.
- **Cost of a wrong call**:
  - **False Positive (refresh recommended for a stable page)**: Wasted editorial budget and human hours (approx. 2-3 hours per page) updating content that does not benefit from it.
  - **False Negative (declining page missed)**: Continued decay of valuable organic traffic, leading to lower clicks, lost conversion opportunities, and eventual client churn.

The Research Question: "Can we use historical search performance (impressions, clicks, average position) and page metadata (word count, age, etc) from a prior 90-day window to predict whether a highly visible page will experience a significant traffic decline over the following 30 days?"

The Unit of Analysis (The Grain): A single pseudonymized content page (content_hash_id) for a specific client (client_hash_id).

The Actionable Output: A prioritized queue of pages ranked by decay risk, complete with specific reason codes (e.g., stale_visible_page, page_one_decay_risk) and recommended actions (e.g., "expand content," "update metadata," "prune").

The Cost of a Wrong Call:
  False Positive (Flagging a healthy page): Waste of editorial time rewriting content that didn't need it.
  False Negative (Missing a decaying page): Permanent loss of organic search traffic and revenue to competitors.

## 3. Quick look at the data (2-3 real numbers)

Based on our exploratory data analysis of the anonymized starter dataset (`data/raw/content_refresh_anonymized.csv`):
1. **High Rate of Decay**: Out of **30,000** total content items, **16,262** pages (**54.21%**) are currently in decline (`trend_direction == 'down'`). This demonstrates that more than half of the active inventory is actively losing traffic.
2. **Aging Content Base**: The average content age is **229.4 days** (median **210 days**). The average days since last update is also **229.4 days**, which indicates that the majority of pages have never been updated since publication.
3. **Massive Traffic at Risk**: Declining pages account for a total of **83,391,371 search impressions** over the trailing 90 days. They have a mean of **5,128 impressions** per page, confirming they are visible pages with real traffic at stake, rather than low-volume noise.

```python
import os
import pandas as pd
import numpy as np

# Find the file path regardless of working directory
path = "data/raw/content_refresh_anonymized.csv"
if not os.path.exists(path):
    path = "../../data/raw/content_refresh_anonymized.csv"
if not os.path.exists(path):
    path = "../data/raw/content_refresh_anonymized.csv"

# Load data
df = pd.read_csv(path)

# Basic stats
total_pages = len(df)
decline_count = (df['trend_direction'] == 'down').sum()
decline_pct = (decline_count / total_pages) * 100

avg_age = df['content_age_days'].mean()
avg_update = df['days_since_last_update'].mean()

declining_df = df[df['trend_direction'] == 'down']
total_impressions_at_risk = declining_df['impressions_90d'].sum()
mean_impressions_declining = declining_df['impressions_90d'].mean()

print(f"Total Pages: {total_pages:,}")
print(f"Declining Pages: {decline_count:,} ({decline_pct:.2f}%)")
print(f"Average Content Age: {avg_age:.1f} days")
print(f"Average Days Since Last Update: {avg_update:.1f} days")
print(f"Total 90d Impressions at Risk: {total_impressions_at_risk:,}")
print(f"Mean 90d Impressions per Declining Page: {mean_impressions_declining:.1f}")
```
